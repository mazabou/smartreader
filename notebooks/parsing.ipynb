Importing data with openers

```python
from openers import *
import pandas as pd
```

```python
articledf = open_metadata()
words = open_words()
col,data = open_data()

articledf['wordcount'] = [sum(v) for v in data]

articledf
```

```python
df_mean = articledf.groupby('source').sum()
df_mean['articlecount']=articledf.groupby('source').size()
df_mean['avg'] = df_mean['wordcount']/df_mean['articlecount']
```

```python
from bokeh.palettes import PuBu
from bokeh.io import show, output_notebook, output_file
from bokeh.models import ColumnDataSource, ranges, LabelSet, HoverTool
from bokeh.plotting import figure
from bokeh.layouts import column, row
from bokeh.layouts import widgetbox
from bokeh.models.widgets import Div
import math
from numpy import pi

output_notebook()
output_file('source.html')

df = articledf.groupby('source').size()
source = ColumnDataSource(dict(x=df.index.tolist(),y=df.values.tolist()))

max_val = max(source.data["y"])
p1 = figure(plot_width=1000, plot_height=400,
        x_axis_label = "Source",
        y_axis_label = "# of articles",
        title="Article sources repartition",
        x_minor_ticks=2,
        x_range = source.data["x"],
        y_range= ranges.Range1d(start=0,end=max_val+5),
        tools=['xpan','box_zoom','reset'])

p1.vbar(source=source,x='x',top='y',bottom=0,width=0.7,color=PuBu[7][2])
p1.xaxis.major_label_orientation = math.pi/2

source2 = ColumnDataSource(dict(x=df_mean.index.tolist(),y=df_mean.avg.tolist()))

max_val = max(source2.data["y"])
p2 = figure(plot_width=1000, plot_height=400,
        x_axis_label = "Source",
        y_axis_label = "Average # of words per article",
        title="Article word count distribution",
        x_minor_ticks=2,
        x_range = source2.data["x"],
        y_range= ranges.Range1d(start=0,end=max_val+100),
        tools=['xpan','box_zoom','reset'])

p2.vbar(source=source2,x='x',top='y',bottom=0,width=0.7,color=PuBu[7][2])
p2.xaxis.major_label_orientation = math.pi/2

df_dist = articledf.groupby('Category').size().sort_values(ascending=False)

distr = [x/float(df_dist.sum()) for x in df_dist.values.tolist()]
percents = [sum(distr[:i]) for i in range(len(distr))] + [sum(distr)]
starts = [p*2*pi for p in percents[:-1]]
ends = [p*2*pi for p in percents[1:]]

# a color for each pie piece
colors = ["red", "green", "blue", "orange", "yellow","grey","white"]

p3 = figure(x_range=(-1.5,1.5), y_range=(-2,2))
p3.wedge(x=0, y=0, radius=1, start_angle=starts, end_angle=ends, color = colors)


layout = row(column(p1, p2),p3)
```

```python
from scipy.sparse import csr_matrix

data_m = [e for v in data for e in v]
col_m = [e for v in col for e in v]
row_m = [e for v in row for e in v]

matrix = csr_matrix((data_m,(row_m, col_m)))

df = pd.DataFrame(matrix.toarray(),columns=colnames)
df['articles'] = articledf.title
df = df.set_index('articles')
```
